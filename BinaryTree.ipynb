```python
db = pd.read_csv('heart.csv')
```

```python
db.dtypes
```

```python
db.rename(columns={
    'Age' : 'edad',
    'Sex'	: 'sexo',
    'ChestPainType'	: 't_dolor_pecho',
    'RestingBP'	    : 'latidos_descanso',
    'Cholesterol'	  : 'colesterol',
    'FastingBS'	    : 'sangre_ayuno',
    'RestingECG'	  : 'electrocardiograma',
    'MaxHR'	: 'frecuencia_maxima',
    'ExerciseAngina': 'angina_ejercicio',
    'Oldpeak'	      : 'pico',
    'ST_Slope'	    : 'pendiente_ejercicio',
    'HeartDisease'  : 'enfermedad_corazon'},
    inplace=True)
```

```python
db['t_dolor_pecho'] = db['t_dolor_pecho'].replace(['TA', 'ATA', 'NAP', 'ASY'],[1, 2, 3, 4]).astype('int64')
db['sexo'] = db['sexo'].replace(['M', 'F'],[1, 2]).astype('int64')
db['electrocardiograma'] = db['electrocardiograma'].replace(['Normal', 'ST', 'LVH'],[1, 2, 3]).astype('int64')
db['angina_ejercicio'] = db['angina_ejercicio'].replace(['N', 'Y'],[0, 1]).astype('int64')
db['pendiente_ejercicio'] = db['pendiente_ejercicio'].replace(['Up', 'Flat', 'Down'],[1, 2, 3]).astype('int64')
```
